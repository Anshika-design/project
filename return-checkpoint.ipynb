```python
import pandas as pd
import numpy as np

# Load your existing dataset
df = pd.read_csv("return_dataset.csv")

# Add dummy return-related columns
reasons = ['Size Issue', 'Defective', 'Changed Mind', 'Wrong Item']
status = ['Pending', 'Approved', 'Rejected']
modes = ['Pickup', 'Drop-off']
dates = pd.date_range(start="2024-01-01", periods=len(df), freq='D')

df['return_date'] = np.random.choice(dates, size=len(df))
df['refund_amount'] = np.random.randint(100, 2000, size=len(df))
df['return_reason'] = np.random.choice(reasons, size=len(df))
df['return_status'] = np.random.choice(status, size=len(df))
df['return_mode'] = np.random.choice(modes, size=len(df))
df['customer_id'] = np.random.randint(1000, 9999, size=len(df))
df['order_id'] = ['ORD' + str(i).zfill(5) for i in range(len(df))]

# Preview
df.head()
```
